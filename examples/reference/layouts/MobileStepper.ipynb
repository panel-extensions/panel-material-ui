The `MobileStepper` displays compact step progress with back/next navigation buttons and a progress indicator. Unlike `Stepper`, it does not display step labels — the number of steps is derived from the number of child objects.

### Parameters:

* **`active`** (`int`): Index of the currently active step.
* **`back_text`** (`str`): Label for the back navigation button.
* **`color`** (`str`): Color of the progress indicator and active dots.
* **`next_text`** (`str`): Label for the next navigation button.
* **`position`** (`Literal["bottom", "static", "top"]`): Positioning of the stepper bar.
* **`variant`** (`Literal["dots", "progress", "text"]`): The type of progress indicator.

---

### Dots Variant

The default variant displays a dot for each step:

```python
pmui.MobileStepper(
    'Configure your campaign name and budget.',
    'Define your target audience.',
    'Upload your creative assets.',
    variant='dots',
)
```

### Progress Bar Variant

The progress variant shows a `LinearProgress` bar:

```python
pmui.MobileStepper(
    'Step 1: Select settings',
    'Step 2: Create ad group',
    'Step 3: Review and submit',
    variant='progress',
)
```

### Text Variant

The text variant displays the current step number and total:

```python
pmui.MobileStepper(
    'Step 1 content',
    'Step 2 content',
    'Step 3 content',
    'Step 4 content',
    'Step 5 content',
    variant='text',
)
```

### Custom Button Text

```python
pmui.MobileStepper(
    'Page 1',
    'Page 2',
    'Page 3',
    back_text='Prev',
    next_text='Forward',
)
```

### Programmatic Navigation

Use `next()`, `back()`, and `reset()` to navigate from Python. Register callbacks with `on_step_change`:

```python
stepper = pmui.MobileStepper(
    'Page 1',
    'Page 2',
    'Page 3',
    variant='progress',
)

stepper.on_step_change(lambda e: print(f'Step changed to {e.new}'))
stepper.next()  # moves to step 1
stepper
```
