```python
import panel as pn
import panel_material_ui as pmui

pn.extension()
```

The `Stepper` layout displays progress through a sequence of logical and numbered steps, useful for wizard-like workflows. It supports horizontal and vertical orientations, linear and non-linear navigation, and step-level states such as completed, optional, disabled, and error.

Like `Tabs`, `Stepper` has a list-like API with methods to `append`, `extend`, `clear`, `insert`, `pop`, `remove` and `__setitem__`, which make it possible to interactively update and modify the steps.

### Parameters:

#### Core

* **`active`** (`int`): Index of the currently active step.
* **`completed`** (`list[int]`): Indices marked as completed. When empty, MUI auto-completes steps before the active step.
* **`disabled`** (`list[int]`): Indices of disabled steps.
* **`dynamic`** (`boolean`): Whether to render step contents only when active.
* **`error`** (`list[int]`): Indices shown in an error state.
* **`icons`** (`list[str]`): Material Icon names per step, replacing the default numbered circles.
* **`non_linear`** (`boolean`): Whether steps are clickable out of order.
* **`optional`** (`list[int]`): Indices marked as optional.

#### Display

* **`alternative_label`** (`boolean`): Whether to place labels below step icons instead of beside them.
* **`color`** (`str`): Color of the active and completed step icons.
* **`connector`** (`boolean`): Whether to display connectors between steps.
* **`orientation`** (`Literal["horizontal", "vertical"]`): The orientation of the stepper.

---

### Basic Usage

The `Stepper` layout accepts tuples of `(label, content)` as positional arguments. The `active` parameter controls which step is currently displayed:

```python
stepper = pmui.Stepper(
    ('Select campaign settings', 'Configure your campaign name, budget, and schedule.'),
    ('Create an ad group', 'Define your target audience and ad placements.'),
    ('Create an ad', 'Upload your creative assets and write ad copy.'),
)

stepper
```

You can get and set the currently `active` step as an integer:

```python
print(stepper.active)
stepper.active = 1
```

### Non-Linear Navigation

By default the stepper is linear — users proceed step by step. Set `non_linear=True` to allow clicking on any step directly:

```python
pmui.Stepper(
    ('Select settings', 'Step 1 content'),
    ('Create group', 'Step 2 content'),
    ('Review', 'Step 3 content'),
    non_linear=True,
)
```

### Vertical Orientation

Use `orientation='vertical'` for a vertical layout. In this mode, step content is rendered inline below each step label:

```python
pmui.Stepper(
    ('Select settings', 'Configure your campaign name, budget, and schedule.'),
    ('Create an ad group', 'Define your target audience and ad placements.'),
    ('Create an ad', 'Upload your creative assets and write ad copy.'),
    orientation='vertical',
    active=1,
)
```

### Alternative Label Placement

Set `alternative_label=True` to place labels below the step icons instead of beside them:

```python
pmui.Stepper(
    ('Select settings', 'Step 1 content'),
    ('Create group', 'Step 2 content'),
    ('Review', 'Step 3 content'),
    alternative_label=True,
    active=1,
)
```

### Step States

Individual steps can be marked with different states using their index:

- `optional` shows an "Optional" label below the step name
- `error` marks a step with an error icon
- `completed` explicitly controls which steps show as completed
- `disabled` prevents interaction with specific steps

```python
pmui.Stepper(
    ('Account details', 'Fill in your account information.'),
    ('Verify email', 'Check your inbox for a verification link.'),
    ('Profile setup', 'Add your name and avatar.'),
    active=1,
    completed=[0],
    error=[1],
    optional=[2],
)
```

### Connectors

Set `connector=False` to remove the lines between steps:

```python
pmui.Stepper(
    ('Step 1', 'Content 1'),
    ('Step 2', 'Content 2'),
    ('Step 3', 'Content 3'),
    connector=False,
)
```

### Icons

Provide a list of [Material Icon](https://fonts.google.com/icons) names to replace the default numbered circles:

```python
pmui.Stepper(
    ('Settings', 'Configure your campaign.'),
    ('Audience', 'Define your target audience.'),
    ('Creative', 'Upload your ad assets.'),
    icons=['settings', 'group_add', 'videocam'],
    active=1,
)
```

### Dynamic Loading

When `dynamic=True`, only the active step's content is rendered. This can improve performance when steps contain expensive components:

```python
pmui.Stepper(
    ('Step 1', pmui.Column(pmui.widgets.Button(label='Action 1'))),
    ('Step 2', pmui.Column(pmui.widgets.Button(label='Action 2'))),
    ('Step 3', pmui.Column(pmui.widgets.Button(label='Action 3'))),
    dynamic=True,
    non_linear=True,
)
```
