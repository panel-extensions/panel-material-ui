The `Badge` generates a small badge to the top-right (by default) of its child element. Badges are commonly used to display notification counts, status indicators, or short labels overlaid on icons, avatars, or buttons.

### Parameters:

* **`anchor_origin`** (`dict`): The anchor position with keys 'vertical' and 'horizontal'.
* **`badge_content`** (`int` or `str`): The content rendered within the badge.
* **`color`** (`str`): The color of the badge.
* **`invisible`** (`bool`): Whether the badge is invisible.
* **`max`** (`int`): Maximum count to display.
* **`object`** (`Viewable`): The child component to wrap with the badge.
* **`overlap`** (`str`): Wrapped shape the badge should overlap.
* **`show_zero`** (`bool`): Whether to display the badge when badge_content is zero.
* **`variant`** (`str`): The variant — 'standard' or 'dot'.

---

### Basic Usage

Apply a badge to a child component:

```python
pmui.Badge(pmui.IconButton(icon="mail"), badge_content=4, color="primary")
```

### Color

Use the `color` parameter to apply different theme colors:

```python
pmui.Row(
    pmui.Badge(pmui.IconButton(icon="mail"), badge_content=4, color="primary"),
    pmui.Badge(pmui.IconButton(icon="mail"), badge_content=4, color="secondary"),
    pmui.Badge(pmui.IconButton(icon="mail"), badge_content=4, color="success"),
    pmui.Badge(pmui.IconButton(icon="mail"), badge_content=4, color="error"),
)
```

### Maximum Value

Use the `max` parameter to cap the displayed value:

```python
pmui.Row(
    pmui.Badge(pmui.IconButton(icon="mail"), badge_content=99, color="secondary"),
    pmui.Badge(pmui.IconButton(icon="mail"), badge_content=100, color="secondary"),
    pmui.Badge(pmui.IconButton(icon="mail"), badge_content=1000, max=999, color="secondary"),
)
```

### Dot Badge

The `dot` variant renders a small dot instead of a count:

```python
pmui.Badge(pmui.IconButton(icon="mail"), variant="dot", color="secondary")
```

### Badge Visibility

Control visibility with the `invisible` parameter:

```python
badge = pmui.Badge(pmui.IconButton(icon="mail"), badge_content=4, color="primary")
toggle = pmui.Switch(value=False, label="Invisible")

pn.bind(lambda v: badge.param.update(invisible=v), toggle.param.value, watch=True)

pmui.Row(badge, toggle)
```

### Badge Alignment

Use the `anchor_origin` parameter to move the badge to any corner:

```python
pmui.Row(
    pmui.Badge(
        pmui.IconButton(icon="mail"), badge_content=1, color="secondary",
        anchor_origin={"vertical": "top", "horizontal": "right"},
    ),
    pmui.Badge(
        pmui.IconButton(icon="mail"), badge_content=2, color="secondary",
        anchor_origin={"vertical": "top", "horizontal": "left"},
    ),
    pmui.Badge(
        pmui.IconButton(icon="mail"), badge_content=3, color="secondary",
        anchor_origin={"vertical": "bottom", "horizontal": "right"},
    ),
    pmui.Badge(
        pmui.IconButton(icon="mail"), badge_content=4, color="secondary",
        anchor_origin={"vertical": "bottom", "horizontal": "left"},
    ),
)
```
